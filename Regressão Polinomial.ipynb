# Regressão Polinomial

### Função sintética

```python
import numpy as np
np.random.seed(42)

m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5*X**3 + 3 * X**2 + X + 0.3 + np.random.randn(m, 1)
```

```python
import matplotlib.pyplot as plt

plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
plt.title("Dados gerados")
plt.show()
```

### Transformação de feature para Linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_features.fit_transform(X)
print(X[0])
print(X_poly[0])
```

### "Fittando" o modelo com Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
print("w0, w1 e w2", lin_reg.intercept_, lin_reg.coef_)
```

### Testando novo ponto

```python
X_new = np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)
plt.plot(X, y, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predição")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-3, 3, 0, 10])
plt.title("Predição gerada")
plt.show()
```
